```python

# Classification tree on Wine data
fig = plt.figure(figsize=(55,50))

clf = tree.DecisionTreeClassifier(criterion='gini',max_depth=None,max_features=6,min_samples_leaf=7)
clf.fit(X_train,y_train)
```

```python
# pick random X observation for demo
#X = wine.data[np.random.randint(0, len(wine.data)),:]
viz = dtreeviz.model(clf,X_train,y_train,
               target_name='Exited',
              #  feature_names=X.columns,
               feature_names=['Geography', 'Gender','Age','Tenure','Balance','NumOfProducts','HasCrCard','IsActiveMember'],
              #  title="Customer Churnclassification",
               class_names=['0', '1'],
              #  histtype='barstacked', # default 
              #  scale=1.2
               )
```
